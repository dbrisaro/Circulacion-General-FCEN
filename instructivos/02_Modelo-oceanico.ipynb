# Modelo de Circulación Oceánica forzada por el viento
Este modelo resuelve la ecuación de vorticidad barotrópica para el océano integrada en la vertical, en su forma no-dimensional, utilizando diferencias finitas.

La clásica ecuación fundamental de vorticidad barotrópica, derivada para un océano forzado por el viento en una cuenca cerrada de profundidad constante $D$, se expresa como:

$$\frac{\partial \varsigma}{\partial t} + J(\psi, \varsigma) + \beta_0 \frac{\partial \psi}{\partial x} = \frac{1}{D} \nabla \times \tau + \tau_b + F,$$ 

donde $\varsigma$ es la vorticidad relativa y $\psi$ la función corriente. La relación entre vorticidad, función corriente y componentes de la velocidad está dada por:

$$\varsigma = \nabla^2\psi$$
$$U = -\frac{\partial \varsigma}{\partial y}; \, V = -\frac{\partial \varsigma}{\partial x}$$

El Jacobiano entre la función corriente y la vorticidad, $J(\psi, \varsigma)$, mide la no-linealidad del problema y se define como:

$$ J(\psi, \varsigma) = \frac{\partial \psi}{\partial x}\frac{\partial \varsigma}{\partial y} - \frac{\partial \varsigma}{\partial x}\frac{\partial \psi}{\partial y}$$

La fricción de fondo ($\tau_b$) se define según Stommel (1948) utilizando un coeficiente de fricción ($K$) que tiene dimensiones de tiempo$^{-1}$:

$$ \tau_b = -K\varsigma $$

La fricción lateral, que puede ser de tipo armónico ($F_h$; Munk, 1950) o biarmónico ($F_{bh}$) (superviscosidad), donde $A_h$ y $B_h$ son coeficientes de viscosidad, se define como:

$$ F_h = A_h \nabla^2 \varsigma $$
$$ F_{bh} = -B_h \nabla^4 \varsigma $$

Por último, $\nabla \times \tau$ es el rotor del viento superficial y $\beta_0 = \frac{\partial f}{\partial y}$ es el gradiente de vorticidad planetaria.

Definiendo escalas características del problema, puede escribirse la ecuación de vorticidad barotrópica en forma no dimensional:

$$(x,y) = L(x',y'); \,(u,v) = U(u',v'); \,t = Tt'$$

entonces, 

$$\frac{\partial \varsigma'}{\partial t} + R_0 J'(\psi', \varsigma') + \frac{\partial \psi'}{\partial x'} = \nabla \times \tau' - \epsilon_s \varsigma' + \epsilon_m \nabla'^2 \varsigma',$$ 

donde $R_0$ es el número de Rossby, $\epsilon_s$ el número de Stommel y $\epsilon_m$ el número de Munk, definidos como:

$$ R_0 = \frac{U}{2 \Omega L}; \, \epsilon_s = \frac{K}{\beta_0 L}; \, \epsilon_m = \frac{A_m}{\beta_0 L^2} $$

donde las escalas típicas del modelo son:

$$ U = \frac{2 \pi \tau_s}{\rho D \beta L}; \, T = \frac{1}{\beta L} $$

Y las escalas de tiempo asociadas a la disipación tienen la forma:

$$ T_f = \frac{1}{K}; \, T_a = \frac{L^2}{A}; \, T_{a4} = \frac{L^4}{A_4}$$

Con esto, se definen nuevamente números adimensionales (parámetros del modelo):


Número de Rossby $R_0 = \dfrac{2 \pi \tau_s}{\rho D \beta^2 L^3}$

Número de Ekman vertical $E_f = \dfrac{K}{\beta L}$

Número de Ekman horizontal $E_{v1} = \dfrac{A}{\beta L^2}$

Número de Ekman bi-armónico $E_{v2} = \dfrac{A_4}{\beta L^5}$

La caracterización de la corriente de borde oeste viene dada por el parámetro $\delta$, que representa la fracción de la longitud de la cuenca que ésta abarca, dando como resultado el ancho de la corriente de borde oeste ($W$). Éste parámetro se obtiene según los efectos a los que esté asociado:

Efectos inerciales: $\delta_i = R^{1/2} \, \rightarrow \, W_i = \delta_i L$

Fricción de fondo: $\delta_i = E_f \, \rightarrow \, W_f = \delta_f L$

Fricción lateral: $\delta_{v1} = E_{v1}^{1/3} \, \rightarrow \, W_{v1} = \delta_{v1} L$

Fricción bi-armónica: $\delta_{v2} = E_{v2}^{1/5} \, \rightarrow \, W = \delta_{v2} L$

Consideraremos las siguientes magnitudes típicas:

| Variable                          | Constante del parámetro |          Magnitud         |
|-----------------------------------|:-----------------------:|:-------------------------:|
| Magnitud del viento               |          $\tau$         |     0.1-0.5 N m$^{-2}$    |
| Longitud de la cuenca             |            L            |          4000 Km          |
| Profundidad                       |            D            |       1000 – 5000 m       |
| Coeficiente de fricción de fondo  |            K            |  1.16x10$^{-7}$ s$^{-1}$  |
| Coeficiente de viscosidad lateral |            A            |  10 - 500 m$^2$ s$^{-1}$  |
| Coeficiente bi-armónico           |          $A_4$          | 108 - 1010 m$^4$ s$^{-1}$ |
| Densidad                          |          $\rho$         |      1025 kg m$^{-3}$     |
| Gradiente meridional de vorticidad planetaria  | $\beta$    |   2x10$^{-11}$ ms$^{-1}$  |
