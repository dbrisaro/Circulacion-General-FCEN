## Definición de funciones 
- Bloque de instrucciones que toma parámetros de entrada y (la mayoría de las veces) devuelve uno de salida
- Las funciones se definen previo a su uso

````python
def suma_de_numeros(a, b):
    return a + b

def suma_de_numeros_pares(a, b):
    if a%2 == 0 and b%2==0:
        c = a + b
    else:
        c = None
    
    return c

a = 0
b = 10
c = suma_de_numeros(a,b)

c = suma_de_numeros_pares(a,b)
print(c)

a = 1
c = suma_de_numeros_pares(a,b)
print(c)
````

## Matplotlib
Librería que permite hace gráficos. Hay varios ejemplos disponibles en [este link](https://matplotlib.org/3.1.1/gallery/index.html)

````python
import matplotlib.pyplot as plt

N = 8
medias_masc = (10, 15, 30, 15, 22, 39, 12, 15)
medias_fem = (25, 32, 34, 20, 25, 11, 9, 10)
std_masc = (2, 3, 4, 1, 2, 1, 2, 1)
std_fem = (3, 5, 2, 3, 3, 1, 1, 3)
````

````python
ind = np.arange(N)    
width = 0.35       

p1 = plt.bar(ind, medias_masc, width, yerr=std_masc, color='lightblue')
p2 = plt.bar(ind, medias_fem, width,
             bottom=medias_masc, yerr=std_fem, color='salmon')

plt.ylabel('Puntaje')
plt.title('Puntajes segun grupo y genero')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Masculino', 'Femenino'))

plt.show()
````

````python
figpath = './'
figname = 'barplot'

fig = plt.figure(figsize=(8,8))
plt.clf()
ax = plt.axes([0.05, 0.05, 0.85, 0.58])
# [xo, yo, width, height]

p1 = ax.bar(ind, medias_masc, width, yerr=std_masc, color='lightblue')
p2 = ax.bar(ind, medias_fem, width,
             bottom=medias_masc, yerr=std_fem, color='salmon')

ax.tick_params(labelsize=12)
ax.set_ylabel('Puntaje', fontsize=12)
ax.set_title('Puntajes segun grupo y genero')
ax.set_xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8'))
ax.set_yticks(np.arange(0, 81, 10))
ax.legend((p1[0], p2[0]), ('Masculino', 'Femenino'), fontsize=12)

fig.savefig(figpath+figname, dpi=400, bbox_inches='tight')
````

````python
# Gráfico de lineas

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 30, dt)
nse1 = np.random.randn(len(t))                 # ruido 1
nse2 = np.random.randn(len(t))                 # ruido 2

# Creamos dos señales mas una parte de ruido
s1 = np.sin(2 * np.pi * 10 * t) + nse1
s2 = np.sin(2 * np.pi * 10 * t) + nse2

fig, axs = plt.subplots(2, 1)
#axs[0].plot(t, s1, t, s2)
axs[0].plot(t, s1)
axs[0].plot(t, s2)

axs[0].set_xlim(0, 2)
axs[0].set_xlabel('Tiempo')
axs[0].set_ylabel('s1 y s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel('Coherencia')
axs[1].set_xlabel('Frecuencia')

fig.tight_layout()
plt.show()
````

````python
#https://matplotlib.org/3.1.0/gallery/color/colormap_reference.html

# Graficamos una imagen con una barra de colores
import matplotlib.cm as cm

delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)

X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

fig, ax = plt.subplots()
im = ax.imshow(Z, interpolation='bilinear', cmap=cm.RdYlGn,
               origin='lower', extent=[-3, 3, -3, 3],
               vmax=abs(Z).max(), vmin=-abs(Z).max())

plt.show()
````

````python
## Quiver
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(np.arange(0, 2 * np.pi, .2), np.arange(0, 2 * np.pi, .2))
U = np.cos(X)
V = np.sin(Y)

fig1, ax1 = plt.subplots()
ax1.set_title('Arrows scale with plot width')
Q = ax1.quiver(X, Y, U, V, units='width')
qk = ax1.quiverkey(Q, 0.9, 0.9, 2, r'$2 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
````

````python
fig2, ax2 = plt.subplots()
ax2.set_title("pivot='mid'; every third arrow; units='inches'")
Q = ax2.quiver(X[::3, ::3], Y[::3, ::3], U[::3, ::3], V[::3, ::3],
               pivot='tail', units='inches')
qk = ax2.quiverkey(Q, 0.9, 0.9, 1, r'$1 \frac{m}{s}$', labelpos='E',
                   coordinates='figure')
ax2.scatter(X[::3, ::3], Y[::3, ::3], color='r', s=5)
````

````python
# Graficamos una imagen con sombreados y contornos

delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)

X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

#Niveles para los contornos
clevs=(-2,-1.5,-1,-0.5,0,0.5,1,1.5,2)

fig, ax = plt.subplots()
im = ax.contourf(X,Y,Z,levels=clevs)
ax.contour(X,Y,Z,levels=clevs,colors='k')

plt.colorbar(im)

plt.show()
````

````python
# Graficamos dos imagenes con sombreados y contornos, con los mismos niveles y una sola colorbar

delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)

X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

Z3 = np.exp(-X**2 - Y**2)
Z4 = np.exp(-(X - 0.6)**2 - (Y - 0.2)**2)
ZZ = -(Z3 - Z4) * 2

#Niveles para los contornos
clevs=[-3,-2,-1.5,-1,-0.5,0,0.5,1,1.5,2,3]

fig, (ax1, ax2) = plt.subplots(ncols=2)
im = ax1.contourf(X,Y,Z,levels=clevs,cmap=cm.BrBG)
ax1.contour(X,Y,Z,levels=clevs,colors='y')

ax2.contourf(X,Y,ZZ,levels=clevs,cmap=cm.BrBG)
ax2.contour(X,Y,ZZ,levels=clevs,colors='y')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(im, cax=cbar_ax)

plt.show()
````

## Archivos binarios de Numpy: .npy y .npz
Formas de guardar arrays
- npy: permite almacenar un único array
- npz: permiten almacenar múltiples arrays.
   Se pueden luego cargar con np.load, que devuelve un diccionario que contiene {nombre: array} como par clave-valor, para cada array.

````python
# Declaro algunos arrays de numpy
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([1, 2])

# Guardo un único array
#np.save('/home/daniufundar/Documents/Docencia/Circulacion/circulacion-2c2022/tutoriales/test.npy', a)
````

````python
data = np.load('./test.npy')
````

````python
# Guardo multiples arrays
np.savez('./test.npz', a=a, b=b)
data = np.load('./test.npz')
a1 = data['a']
b1 = data['b']
data.close()
````
