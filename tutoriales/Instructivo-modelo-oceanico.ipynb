# Instructivo para el uso del modelo oceánico
Dentro del directorio de trabajo ```~/circulacion-2c2020/modulo_oceano/``` van a poder encontrar el material para realizar las prácticas de circulación, donde ejecutaremos un modelo numérico de circulación oceánica.

El modelo se encuentra dentro del directorio ```~/circulacion-2c2020/modulo_oceano/modeloQG/``` y las salidas del modelo se guardan en ```~/circulacion-2c2020/modulo_oceano/modeloQG/out_tmp``` $\rightarrow$ ojo!! Cada vez que ejecuten el modelo, asegurense que la carpeta está vacía. 

El código fuente del modelo está escrito en Fortran, en el archivo $\rightarrow$ ```QG_barotrop.f```

Los parámetros del modelo están en un archivo que modificaremos para cada uno de los ejercicios de la práctica $\rightarrow$ ```QG_param.dat``` 

El archivo ```QG_param.dat``` tiene la forma:

	im=202           	% number of grid points in the zonal direction
	jm=102           	% number of grid points in the meridional direction
	ds=0.05          	% grid step
	dt=0.05          	% time step
	Ro=0.0           	% Rossby number (measures non-linearity of the flow)
	eps=0.3          	% non-dimensional coefficient representing bottom friction
	Ah=0.0           	% non-dimensional coeff. of horizontal Laplacian mixing
	Bh=0.0           	% non-dimensional coeff. of horizontal bi-harmonic mixing
	gamma=0.0        	% coeff. of "intermediate slipping" used as boundary cond.
	nst=1            	% start time step number
	nend=2000        	% end time step number
	nlpt=100         	% frequency (time steps) for saving output
	MCF=0            	% matrix (0) or column (1) output
	ncrit=4000       	% number of steps allowed to do the relaxation (sub. helm)
	pcrit=0.1        	% criterium to stop the relaxation
	BFP=1            	% Beta (BFP=1) or F plane (BFP=0)
	GYR=1            	% Simple Gyre (GYR=1) or Double Gyre (GYR=2)
	HEM=-1           	% North Hemisphere Gyre (HEM=1) or South Hemisphere Gyre (HEM=-1)

### Compilación del código

Para la compilación del código usamos la siguiente linea de comando en la terminal:

```gfortran archivo fuente -o archivo ejecutable```

En nuestro caso, el archivo fuente es ```QG_barotrop.f``` y como convención, llamaremos ```QG``` al archivo ejecutable:

```gfortran QG_barotrop.f -o QG```

IMPORTANTE: si ```gfortran``` no está instalado, podés instalarlo siguiente los siguientes pasos:

``` cd ~/circulacion-2c2020/fortran-setup/ ```

``` bash instalacion.sh ```

---
### Ejecución del modelo

Recordar que debe existir el directorio out_tmp dentro del directorio de ejecución del modelo:
```~/circulacion-2c2020/modulo_oceano/modeloQG/out_tmp```

Para correr el modelo, usamos el nombre que le dimos al archivo ejecutable cuando compilamos con un ```./``` adelante en la terminal, es decir:
```./QG```

Archivos de resultados (salidas)
1. Función corriente: _psi##.dat_
2. Vorticidad: _vor##.dat_
3. Variables en el punto central del dominio: _QG_diag.dat_
4. Rotor de la tensión del viento: _QG_wind_stress.dat_

Todos son archivos ASCII, y pueden visualizarse con cualquier editor de texto, por ejemplo:
```nano QG_diag.dat```  

(o  gedit, vim, vi, etc.)

### Cargar la salida del modelo
Para cargar en python la salida del modelo, vamos a usar el programa ```cargar_salida_QG.py```

Se obtienen:
1. _QG_diag_ $\rightarrow$ tiene cuatro columnas, en este orden: tiempo, función corriente ($\psi$), vorticidad en el punto central del dominio ($\varsigma$) y energía cinética (esta última en todo el dominio).
2. _psi_adim_ $\rightarrow$ tiempo x nodos meridionales x nodos zonales tiene todos los campos de función corriente ($\psi$) en todos los tiempos que se le pidió al modelo que sacara las salidas
3. _psi_adimF_ $\rightarrow$ función corriente ($\psi$) en el último tiempo
4. _vor_adim_ $\rightarrow$ idem psi_adim pero con la vorticidad ($\varsigma$)
5. _vor_adimF_ $\rightarrow$ vorticidad relativa en el último tiempo
6. _QG_curlw_ $\rightarrow$ rotor del viento utilizado en la simulación
