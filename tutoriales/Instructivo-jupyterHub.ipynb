## Instructivo del uso del jupyterHub

1. Todos los miembros de circulacin tienen acceso a jupyterHub a travs de un usuario y contrasenia que sern dados en clase. 

2. Luego de loguearse en el Hub elegir la opcin "Circulacin Enviroment"

3. Una vez dentro del Hub podrn ver un navegador de carpetas a la izquierda. Este panel se puede ocultar o mostrar usando el atajo Ctrl + B

4. Una vez que ingresan el Hub debera verse de la siguiente manera:



5. Es probable que la carpeta *circulacion-2c2020* no aparezca dentro de su usuario, para ello tenemos que clonarla de GitHub:
    
    A. Abrir una terminal y ejecutar el siguiente comando 
    
    ```git clone https://github.com/dbrisaro/circulacion-2c2020.git```
    
    B. El repositorio de github se ir actualizando a medida que avanza el cuatrimestre, para poder tener la carpeta *circulacion-2c2020* en su ltima versin pueden hacer un update a travs de los comandos  
    
    ```cd ~/circulacion-2c2020```
    
    ```git pull```
    
    o bien se puede remover el directorio y luego volver a clonarlo:
    
    ```rm -rf ~/circulacion-2c2020```

6. Luego de esto, la carpeta *circulacion-2c2020* aparecer en el panel de la izquierda. Dentro de ella encontrarn las guas prcticas de ejercicios, tutoriales y materiales varios para llevar adelante las prcticas
