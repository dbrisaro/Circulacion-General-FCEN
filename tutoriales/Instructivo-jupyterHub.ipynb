## Instructivo del uso del jupyterHub

1. Todos los miembros de circulación tienen acceso a jupyterHub a través de un usuario y contrasenia que serán dados en clase. 

2. Luego de loguearse en el Hub elegir la opción "Circulación Enviroment"

3. Una vez dentro del Hub podrán ver un navegador de carpetas a la izquierda. Este panel se puede ocultar o mostrar usando el atajo Ctrl + B

4. Una vez que ingresan el Hub debería verse de la siguiente manera:



5. Es probable que la carpeta *circulacion-2c2020* no aparezca dentro de su usuario, para ello tenemos que clonarla de GitHub:
    
    A. Abrir una terminal y ejecutar el siguiente comando 
    
    ```git clone https://github.com/dbrisaro/circulacion-2c2020.git```
    
    B. El repositorio de github se irá actualizando a medida que avanza el cuatrimestre, para poder tener la carpeta *circulacion-2c2020* en su última versión pueden hacer un update a través de los comandos  
    
    ```cd ~/circulacion-2c2020```
    
    ```git pull```
    
    o bien se puede remover el directorio y luego volver a clonarlo:
    
    ```rm -rf ~/circulacion-2c2020```

6. Luego de esto, la carpeta *circulacion-2c2020* aparecerá en el panel de la izquierda. Dentro de ella encontrarán las guías prácticas de ejercicios, tutoriales y materiales varios para llevar adelante las prácticas


____

### Puntos importantes

- Pueden crear carpetas y subir archivos dentro de su usuario del Hub y los docentes podemos acceder a ellas en modo lectura, en caso que lo necesiten. 

- NO pongan su carpeta personal dentro de la carpeta *circulacion-2c2020*

- Si bien el Hub guardará la información de sus archivos, recuerden hacer un backup al terminar el día, ante cualquier eventualidad.

- Cada vez que terminen de usar el Hub no olviden de ir a *File* $\rightarrow$ *Log out*
